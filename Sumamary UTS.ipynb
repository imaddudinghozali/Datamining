```python
import pandas as pd #artinya memanggil library pandas dan memberi nama singkat "pd" supaya lebih mudah digunakan.
```

```python
Data1 = pd.read_csv ("Benign Traffic.csv") #digunakan untuk membaca file CSV bernama "Benign Traffic.csv" dan menyimpannya ke dalam variabel Data1 sebagai sebuah DataFrame pandas.
```

```python
Data1 #Jika Anda mengetikkan Data1 di Python, maka akan menampilkan isi data tersebut dalam bentuk tabel.
```

```python
data2 = pd.read_csv("DDoS ICMP Flood.csv") #digunakan untuk membaca file CSV bernama "DDoS ICMP Flood.csv" dan menyimpannya ke dalam variabel data2 sebagai DataFrame pandas.
```

```python
data2 ##Jika Anda mengetikkan Data1 di Python, maka akan menampilkan isi data tersebut dalam bentuk tabel.
```

```python
data3 = pd.read_csv("DDoS UDP Flood.csv") #digunakan untuk membaca file dan menyimpannya dalam variabel data3
```

```python
data3 #menampilkan file tadi yang disimpen dalam variabe data3
```

```python
gabungin =pd.concat([Data1,data2,data3],ignore_index=True) #digunakan untuk menggabungkan tiga DataFrame (Data1, data2, dan data3) menjadi satu DataFrame besar. Parameter ignore_index=True membuat indeks baru yang berurutan dari 0, sehingga indeks lama dihapus dan diganti.
```

```python
gabungin #gabungin adalah variabel yang berisi hasil penggabungan data dari beberapa DataFrame (misalnya Data1, data2, dan data3) menjadi satu tabel besar. 
```

```python
gabungin.columns.values #Perintah gabungin.columns.values digunakan untuk menampilkan daftar nama kolom yang ada di DataFrame gabungin.
```

```python
x = gabungin.iloc[:,7: 76] #Perintah x = gabungin.iloc[:, 7:76] digunakan untuk mengambil subset kolom dari DataFrame gabungin, yaitu kolom dari indeks ke-7 sampai indeks ke-75 (karena indeks akhir di Python eksklusif).
```

```python
x #menampilkan data yang ada didalam variabel x
```

```python
y = gabungin.iloc[:,83: 84 # berfungsi untuk mengambil kolom ke-83 dari DataFrame gabungin sebagai subset data.
```

```python
y #menampilkan data yang ada divaribel y
```

```python
from sklearn.model_selection import train_test_split #Perintah from sklearn.model_selection import train_test_split digunakan untuk mengimpor fungsi train_test_split dari library scikit-learn. Fungsi ini sangat berguna untuk membagi dataset menjadi dua bagian, biasanya data latih (training) dan data uji (testing), agar model machine learning bisa dilatih dan diuji secara terpisah.
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42) 
#berfungsi untuk membagi dataset x dan y menjadi data latih dan data uji.

#test_size=0.2 berarti 20% data akan digunakan sebagai data uji, dan 80% sisanya untuk data latih.
#random_state=42 digunakan agar pembagian data ini bisa direproduksi (hasilnya sama setiap kali dijalankan).
#Jadi, setelah perintah ini, Anda punya empat variabel:

#x_train dan y_train untuk melatih model
#x_test dan y_test untuk menguji performa model.
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
#digunakan untuk mengimpor modul pohon keputusan (decision tree) dari library scikit-learn. Dengan ini, Anda bisa membuat model klasifikasi menggunakan algoritma Decision Tree.
```

```python
alya = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
alya.fit(x_train,y_train)
# membuat model pohon keputusan dengan kriteria pengukuran entropi dan metode pemilihan split secara acak, lalu melatih model tersebut menggunakan data latih x_train dan y_train.
```

```python
y_pred = alya.predict(x_test)
#digunakan untuk memprediksi hasil atau kelas dari data uji x_test menggunakan model Decision Tree yang sudah dilatih (alya). Hasil prediksi ini disimpan di variabel y_pred.
```

```python
y_pred #menampilkan  data yang ada di variabel y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
#digunakan untuk menghitung akurasi model, yaitu seberapa tepat prediksi model (y_pred) dibandingkan dengan data sebenarnya (y_test). Nilai accuracy yang dihasilkan biasanya berupa angka antara 0 sampai 1, di mana semakin mendekati 1 berarti model semakin akurat.
```

```python
accuracy #menampilkan data yang ada di variabel accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
#digunakan untuk mengimpor dua library penting di Python:

#matplotlib.pyplot (disingkat plt) untuk membuat grafik dan visualisasi data,
#numpy (disingkat np) untuk operasi matematika dan manipulasi array yang efisien.
#Biasanya kedua library ini sering dipakai bersama untuk analisis data dan visualisasi hasil, misalnya menampilkan grafik akurasi model atau distribusi data.
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(alya, feature_names = x.columns.values, class_names = np.array([ 'Benign Traffic','DDos ICMP Flood','DDoS UDP Flood']), filled = True)
plt.show()

#Perintah ini digunakan untuk menampilkan visualisasi pohon keputusan (decision tree) yang sudah Anda buat dengan model alya.

#fig = plt.figure(figsize=(10, 7)) membuat ukuran gambar grafik sebesar 10x7 inci.
#tree.plot_tree() menggambar struktur pohon keputusan dengan nama fitur dari x.columns.values dan nama kelas yang Anda tentukan (Benign Traffic, DDoS ICMP Flood, DDoS UDP Flood).
#Parameter filled=True membuat setiap node pohon diwarnai sesuai kelasnya.
#plt.show() menampilkan grafik tersebut.
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'Benign Traffic','DDos ICMP Flood','DDoS UDP Flood'])
#import seaborn as lol mengimpor library seaborn untuk visualisasi data, tapi dengan alias lol (biasanya alias yang umum adalah sns). Seaborn membantu membuat grafik yang lebih menarik dan informatif.
#from sklearn import metrics mengimpor modul metrics dari scikit-learn yang berisi fungsi-fungsi untuk mengevaluasi performa model machine learning, seperti akurasi, precision, recall, confusion matrix, dan lain-lain.
#label = np.array([...]) membuat sebuah array NumPy yang berisi nama-nama kelas yang digunakan dalam dataset Anda, misalnya jenis-jenis traffic atau serangan.
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_pred) #membuat matriks kebingungan yang membandingkan hasil prediksi (y_pred) dengan data asli (y_test).

plt.figure(figsize=(10, 10)) #mengatur ukuran gambar heatmap.
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label) #membuat heatmap dengan nilai angka di setiap kotak, serta label sumbu x dan y sesuai nama kelas.
plt.xlabel('Prediksi') #memberi label pada sumbu x dan y.
plt.ylabel('Fakta') #memberi label pada sumbu x dan y.
plt.show() #menampilkan grafik.
```
